# An√°lise explorat√≥ria de dados

## Qual o valor m√©dio de aluguel por tipo de im√≥vel?

```python
display(dados.head(8))
```

```python
# Valor m√©dio de todos im√≥veis
print(dados['Valor'].mean())
```

```python
# Agrupando os dados de acordo com a coluna 'Tipo' e calculando a m√©dia deles
display(dados.groupby('Tipo').mean(numeric_only=True))
```

```python
# Selecionando apenas os alugueis (coluna 'Valor') e calculando a m√©dia dos mesmos por tipo
# Repare que obtemos uma serie do dataframe (DataframeGroupby no caso)

media_alugueis = dados.groupby('Tipo')['Valor'].mean()
display(media_alugueis)
```

```python
# Melhorando a visualiza√ß√£o transformando em um dataframe e ordenando de forma crescente
"""Nesse caso, para obter um dataframe, basta passar a lista ['Valor'] em vez da string 'Valor'.
   Pois o m√©todo groupby entende que voc√™ deseja passar v√°rias colunas e assim te devolve um dataframe
   mesmo que seja apenas uma coluna nessa lista"""

display(dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor'))
```

```python
# Plotando gr√°fico...
media_alugueis = dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
media_alugueis.plot(kind='barh', figsize=(14, 10), color='purple', xlabel='Valor m√©dio')
```

## Removendo os im√≥veis comerciais

```python
# Todos os tipos
display(dados.Tipo.unique())
```

```python
# Dentro dos tipos esses s√£o considerados comerciais por n√≥s
imoveis_comerciais = ('Conjunto Comercial/Sala',
                      'Pr√©dio Inteiro', 'Loja/Sal√£o',
                      'Galp√£o/Dep√≥sito/Armaz√©m',
                      'Casa Comercial', 'Terreno Padr√£o',
                      'Loja Shopping/ Ct Comercial',
                      'Box/Garagem', 'Ch√°cara',
                      'Loteamento/Condom√≠nio', 'S√≠tio',
                      'Pousada/Chal√©', 'Hotel', 'Ind√∫stria')
```

```python
# Consultando apenas os im√≥veis comerciais
display(dados.query('@imoveis_comerciais in Tipo'))
```

```python
# Agora removendo os im√≥veis comerciais na consulta
display(dados.query('@imoveis_comerciais not in Tipo'))
```

```python
# Salvando os im√≥veis residenciais
df = dados.query('@imoveis_comerciais not in Tipo')

df.head() # Tinha um conjunto comercial no √≠ndice 2, n√£o tem mais!
```

```python
# Conferindo mais uma vez...
display(df.Tipo.unique())
```

```python
# Plotando novamente um gr√°fico, apenas com os im√≥veis residenciais
df_preco_tipo = df.groupby('Tipo').mean()[['Valor']].sort_values('Valor')

df_preco_tipo.plot(kind='barh', figsize=(14, 10), color ='purple', xlabel='Valor m√©dio')
```

## Qual o porcentual de cada tipo de im√≥vel na nossa base de dados?

```python
# Contando o quanto temos de cada tipo de im√≥vel (frequ√™ncia absoluta de cada)
display(df.Tipo.value_counts())
```

```python
# Normalizando (frequ√™ncia relativa de cada), convertendo em um dataframe e ordenando
display(df.Tipo.value_counts(normalize=True).to_frame().sort_values('Tipo'))
```

```python
# Renomeando a coluna Tipo no dataframe anterior para melhorar a visualiza√ß√£o
df_tipo = df.Tipo.value_counts(normalize=True).to_frame().sort_values('Tipo')
df_tipo.rename(columns={'Tipo': 'Porcentagens'}, inplace=True)
display(df_tipo)
```

```python
# Salvando e plotando...
df_normalizado_tipo = df.Tipo.value_counts(normalize=True).to_frame().sort_values('Tipo')

df_normalizado_tipo.plot(kind='bar', figsize=(14,10), color='green', xlabel='Tipos', ylabel='Porcentual')
```
