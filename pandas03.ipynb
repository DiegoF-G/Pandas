# Tratando, filtrando, alterando e salvando alterações em uma base de dados

```python
import pandas as pd
```

Importando novamente o arquivo dataset [**alunos.csv**](https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv), usado no arquivo pandas01.ipynb

```python
# Importando novamente os dados
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv'
dados = pd.read_csv(url)
display(dados)
```

## Verificando se a base de dados possui dados nulos e caso tenha, vamos realizar o tratamento desses dados nulos

```python
# Tem dados nulos?
display(dados.isnull().sum())
```

```python
# Provavelmente os alunos sem nota tiraram 0, então basta preencher com 0 essas notas...
dados.fillna(0, inplace=True)
display(dados['Notas'])
```

## Os alunos "Alice" e "Carlos" não fazem mais parte da turma e portanto vamos remove-los

```python
# Consultando as linhas (índices) dos alunos em questão...
alunos_a_remover = dados.query('Nome == "Alice" | Nome == "Carlos"').index
display(alunos_a_remover)
```

```python
# Removendo e confirmando a remoção
dados.drop(alunos_a_remover, inplace=True)
display(dados['Nome'])
```

## Filtrando apenas os alunos que foram aprovados

```python
aprovados = dados['Aprovado'] == True
```

```python
alunos_aprovados = dados[aprovados]
display(alunos_aprovados)
```

## Salvando o DataFrame "alunos_aprovados" em um arquivo csv chamado "alunos_aprovados.csv"

```python
alunos_aprovados.to_csv('alunos_aprovados.csv', index=False)
```

```python
# Verificando se foi salvo corretamente
display(pd.read_csv('alunos_aprovados.csv'))
```

## Ao conferir as notas dos alunos aprovados, notamos que algumas notas estavam incorretas. Os alunos que tiraram nota 7.0, na verdade tinham um ponto extra que não foi contabilizado. Sendo assim, vamos substituir as notas 7.0 da base de dados por 8.0

```python
# Para ser aprovado é necessário ter nota maior ou igual a 6, sendo assim podemos alterar direto no dataframe anteriror
alunos_aprovados.replace(7.0, 8.0, inplace=True)
alunos_aprovados.to_csv('alunos_aprovados_corrigido.csv', index=False)
display(alunos_aprovados)
```
