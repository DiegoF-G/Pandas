# Criando colunas

```python
import pandas as pd
```

Importando novamente o arquivo dataset [**alunos.csv**](https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv), usado no arquivo pandas01.ipynb e pandas03.ipynb

```python
# Importando novamente os dados
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv'
dados = pd.read_csv(url)
display(dados)
```

## Os alunos participaram de uma atividade extracurricular e ganharam pontos extras. Esses pontos extras correspondem a 40% da nota atual de cada um deles.

Vamos criar uma coluna que contenha os pontos extras de cada aluno, ou seja, 40% da nota atual deles.

```python
dados = dados.assign(Pontos_extras = dados['Notas']*0.4)
display(dados)
```

## Vamos criar mais uma coluna, que possua as notas de cada aluno somada com os pontos extras.

```python
dados['Notas_finais'] = dados['Notas'] + dados['Pontos_extras']

# Ajustandos as notas maiores que 10, o ideal é a nota máxima ser 10...
dados['Notas_finais'] = dados['Notas_finais'].apply(lambda nota: 10 if nota > 10 else nota)

display(dados)
```

## Como houve uma pontuação extra, alguns alunos que não tinham sido aprovados antes podem ter sido aprovados agora. Portanto, vamos criar uma coluna com os seguintes valores:

- True: caso o aluno esteja aprovado (nota final deve ser maior ou igual a 6);
- False: caso o aluno esteja reprovado (nota final deve ser menor que 6).

```python
dados['Aprovado_final'] = dados['Notas_finais'].apply(lambda nota: True if nota >= 6 else False)
display(dados)
```

## Fazendo uma seleção e verificando quais alunos não tinham sido aprovados anteriormente, mas foram aprovados após a soma dos pontos extras.

```python
# Poderiamos usar o método query também...
aprovados_posteriormente = (dados['Aprovado'] == False) & (dados['Aprovado_final'] == True)
selecao_aprovados_posteriormente = dados[aprovados_posteriormente]

display(selecao_aprovados_posteriormente)
```
