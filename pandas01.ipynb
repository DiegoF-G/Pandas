```python
import pandas as pd
```

#Importando e armazenando dados

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv'
dados = pd.read_csv(url)
display(dados)
```

#Visualizando parte dos dados

```python
print(dados.head(7)) # 7 primeiros
print(dados.tail(10)) # 10 últimos
```

#Verificando tamanho da base de dados

```python
print(dados.shape)
```

#Explorando um pouco os dados

```python
display(dados.info())
```

```python
display(dados[['Nome', 'Idade', 'Aprovado']])
print()
display(dados[['Notas']]) # Sem os colchetes "[]" na string 'Notas, obtemos uma Serie pandas em vez de um dataframe
```

#Resumo estatístico dos dados
Analisando têndencias, dispersões, máximos/mínimos

```python
# Método describe
display(dados.describe())

# Analisando apenas estatísticas das notas
print()
display(dados.describe()[['Notas']])
```
