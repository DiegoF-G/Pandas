#**Nesse arquivo irei explorar e extrair mais informações do dataset sendo estudado no arquivo "projeto_imobiliaria.ipynb"**

```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url, sep=';')
```

```python
dados.head(10)
```

# Calculando a média de quartos por apartamento

```python
media_quartos_ap = dados.query('Tipo=="Apartamento"')['Quartos'].mean()
print(media_quartos_ap) # 2 quartos por ap mais ou menos
```

# Conferindo quantos bairros únicos existem na nossa base de dados

```python
# Método nunique
quantidade_bairros = dados['Bairro'].nunique()
display(quantidade_bairros)
```

# Analisando quais bairros possuem a média de valor de aluguel mais elevadas para apartamentos

```python
ap = dados.query('Tipo=="Apartamento"')
mais_caros = ap.groupby('Bairro')[['Valor']].mean().sort_values('Valor', ascending=False)
display(mais_caros.head(6)) # Os 6 mais caros apartamentos da base de dados
```

# Plotando gráfico de barras horizontais com os 5 bairros com as médias mais elevadas de valor de aluguel de apartamentos.

```python
display(mais_caros.head(5).plot(kind='barh', figsize=(7,5), xlabel='Aluguel médio', color="yellow"))
```
