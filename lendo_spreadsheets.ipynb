# Lendo dados do Google Planilhas

Vamos ler uma página de uma planilha direto do Google Planilhas como um DataFrame usando o Pandas, a planilha é referente a dados de emissões de carbono *per capita* de vários países de 1990 a 2019.

Para isso precisamos criar uma url com parâmetros adequados tendo como base a url da página da planilha do Google Planilhas...

```python
import pandas as pd
```

1. Vamos primeiro olhar e obter o "id", da página da planilha, na url da mesma aberta no Google Planilhas.

Url em questão: https://docs.google.com/spreadsheets/d/1pvBoLyX8kP0TjtUbadVMGdTl4yzm6bHMThhPiqCVtpw/edit#gid=1214654498


```python
# O "id" está entre "d/" e "/edit" na url
sheet_id = '1pvBoLyX8kP0TjtUbadVMGdTl4yzm6bHMThhPiqCVtpw'
```

2. Em seguida vamos montar, a partir da url base https://docs.google.com/spreadsheets/d/, a url para acessar através do Pandas a planilha no Google Planilhas. Começando com o anexo do "sheet_id" obtido anteriormente no final da url base mencionada que vamos chamar de "sheet_url".

```python
# Interpolando no final da url base mencionada o "sheet_id"
sheet_url = f'https://docs.google.com/spreadsheets/d/{sheet_id}'
```

3. Anexando o parâmetro "/gviz/tq" na "sheet_url", para usar a API do google de visualização que vai conectar o Google Colab e o Google planilhas.

```python
google_API = '/gviz/tq'
sheet_url=f'https://docs.google.com/spreadsheets/d/{sheet_id}{google_API}'
```

4. O próximo parâmetro será para termos como retorno um arquivo csv na forma de página de planilha. Esse é um recurso da API do Google usada.

```python
format_csv_sheet= '?tqx=out:csv&sheet'
sheet_url=f'https://docs.google.com/spreadsheets/d/{sheet_id}{google_API}{format_csv_sheet}'
```

Agora nossa url está pronta para ser usada pelo read_csv (não a read_excel, mas poderia ser...) do Pandas :)

```python
# Com a sheet_url pronta, vamos obter nosso dataframe que é uma planilha excel la no Google Planilhas!
dados = pd.read_csv(sheet_url)
display(dados)
```
